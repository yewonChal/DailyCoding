```python
# 실무예시
# 함수 선언부
def add_data(friend):
    # 1단계 : 빈칸 확보
    katok.append(None) # katok[다현, None] 생성
    klen = len(katok) # 1
    # 2단계 : 친구 대입
    katok[klen-1] = friend # katok[다현, 정현, ...]

def insert_data(position, friend):
    # 1단계 : 빈칸 확보
    katok.append(None)
    klen = len(katok)
    # 2단계 : 마지막 칸부터 3등까지 한 칸씩 뒤로 이동
    for i in range(klen-1, position, -1):
        katok[i] = katok[i-1]
        katok[i-1] = None
    # 3단계 : 친구 대입
    katok[position] = friend

def delete_data(position):
    # 1단계 : 그 위치 지우기
    katok[position] = None
    klen = len(katok)
    # 2단계 : 지운 위치 다음 칸부터, 마지막 친구까지 앞으로 한 칸
    for i in range(position+1, klen, 1):
        katok[i-1] = katok[i]
        katok[i] = None
    # 3단계 : 마지막 칸 제거
    del(katok[klen-1])
    
# 전역 변수부
katok = []

# 메인 코드부
add_data('다현')
add_data('정연')
add_data('쯔위')
add_data('사나')
add_data('지효')
print(katok)
add_data('모모')
print(katok)

insert_data(3, '미나')
print(katok)

delete_data(4)
print(katok)
```

```python
# 단순 연결 리스트 p.17
## 함수부
class Node(): # 자료형 생성
    def __init__(self):
        self.data = None
        self.link = None

def printNodes(start):
    head = start
    current = head
    print(current.data, end=' ')
    while (current.link != None):
        current = current.link
        print(current.data, end=' ')
    print()

def insertNode(findData, insertData):
    global memory, head, current, pre
    # Case 1 : 하필 머리 앞에 삽입(다현, 화사)
    if (findData == head.data):
        node = Node()
        node.data = insertData
        node.link = head
        head = node
        memory.append(node) # head 생성
        return 
    # Case 2 : 중간 노드 앞에 삽입(사나, 솔라)
    current = head
    while (current.link != None):
        pre = current
        current = current.link
        if (current.data == findData):
            node = Node()
            node.data = insertData
            node.link = current
            pre.link = node
            memory.append(node) # head 생성
            return
    # Case 3 : 마지막 노드에 삽입(없는 노드 앞에 삽입(재남, 문별))
    node = Node()
    node.data = insertData
    current.link = node
    return

def deleteNode(deleteData):
    global memory, head, current, pre
    # Case 1 : 하필 삭제할 데이터가 머리일 때(다현)
    if (head.data == deleteData):
        current = head
        head = head.link
        del (current)
        return
    # Case 2 : 중간 노드 삭제(쯔위)
    current = head
    while (current.link != None):
        pre = current
        current = current.link # pre와 current가 같이 가게 됨
        if (current.data == deleteData):
            pre.link = current.link
            del(current)
            return
    # Case 3 : 지울 노드가 존재하지 않을 때(없는 데이터를 삭제할 때(재남))
    return

def findNode(fData):
    global memory, head, current, pre
    current = head
    if (current.data == fData):
        return current
    while (current.link != None):
        current = current.link
        if (current.data == fData):
            return current
    return Node()

## 변수부
memory = []
head, current, pre = None, None, None
dataArray = ['다현', '정연', '쯔위', '사나' ,'지효'] # 다양한 종류와 크기의 데이터를 저장할 수 있어야 한다.

## 메인
# 연결 리스트 생성(개수에 무관함)
node = Node()
node.data = dataArray[0]
head = node
memory.append(node) # head 생성

# head를 제외하고 반복문 처리
for data in dataArray[1:] : # dataArray에서 정연부터 끝까지
    pre = node
    node = Node()
    node.data = data
    pre.link = node
    memory.append(node) # head 생성

printNodes(head)

### 노드 삽입
# 노드 삽입(첫 번째 노드 삽입)(헤드가 바뀌는 상황) 
# insertNode('다현', '화사') # 찾을 데이터, 삽입형 데이터
# printNodes(head)

# 노드 삽입(중간에 삽입)
# insertNode('사나', '솔라') # 찾을 데이터, 삽입형 데이터
# printNodes(head)

# 마지막 노드 삽입
# insertNode('재남', '문별') # 찾을 데이터, 삽입형 데이터
# printNodes(head)

### 노드 삭제
# 노드 삭제(첫 번째 노드 삭제)(헤드가 바뀌는 상황)
# deleteNode('다현') # 삭제할 데이터
# printNodes(head)

# 노드 삭제(중간 노드 삭제)
# deleteNode('쯔위') # 삭제할 데이터
# printNodes(head)

# 노드 삭제(지울 노드가 실제로 존재하지 않을 때)(없는 데이터를 삭제)
# deleteNode('재남') # 삭제할 데이터
# printNodes(head)

# 특정 데이터를 찾았을 때, 또다른 데이터도 함께 다올 수 있도록 설계
findData = '사나'
fNode = findNode(findData)
print(fNode.data, '뮤비가 플레이 됩니다.')
```

```python
## 함수
# 오버플로우 방지 코드
def isStackFull():
    global SIZE, stack, top
    if (top >= SIZE -1):
        return True
    else:
        return False

# 데이터 삽입(push)
def push(data):
    global SIZE, stack, top
    if (isStackFull()):
        print("'", data, "'", '-> 오버플로우 발생 방지(여분 자리가 없습니다.)', '\n')
        return
    top += 1
    stack[top] = data

def isStackEmpty():
    global SIZE, stack, top
    if (top <= -1):
        return True
    else:
        return False

def pop():
    global SIZE, stack, top
    if (isStackEmpty()):
        print('언더플로우 발생 방지(들어가있는 데이터가 없습니다.)', '\n')
        return
    data = stack[top]
    stack[top] = None
    top -= 1
    return data

def peak():
    global SIZE, stack, top
    if (isStackEmpty()):
        print('들어가있는 데이터가 없습니다.', '\n')
        return
    return stack[top]
    
## 변수
SIZE = 5
stack = [None for _ in range(SIZE)]
top = -1

## 메인
push('커피')
push('녹차')
# push('주스')
# push('콜라')
# push('맥주')
# push('사이다')

print('바닥 | ', stack)

retData = pop()
print('팝 데이터 --> ', retData)

print('다음 예정 : ', peak())

retData = pop()
print('팝 데이터 --> ', retData)

retData = pop()
print('팝 데이터 --> ', retData)
print('바닥 | ', stack)
```
