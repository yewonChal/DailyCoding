2. 사람 데이터를 이용해 풀어보기
- 도시별 평균 나이 구하기
- 가장 많은 사람이 살고 있는 도시
- 연봉을 추가해서 랜덤으로 할당 (50,000 ~ 300,000)
- 연봉 순서대로 정렬한 후 연봉의 1등과 꼴등 출력해보기

```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Name': ['Person_' + str(i) for i in range(1,1001)],
    'Age' : np.random.randint(20,60,1000),
    'Salary' : np.random.randint(50000, 300000, 1000),
    'City' : np.random.choice(["Seoul", "Busan", "Daegu", "Incheon", "Gwangju", "Daejon", "Ulsan"], 1000)
})

print(data)
print()

print("도시별 평균 나이 : ")
# avg = data.groupby("City")["Age"].mean()
# print(avg, '\n')
# gorupby 대신 쓸 수 있는 용어 : df['City']
# city에 속하는 age 출력

city_age = {} # 딕셔너리 생성
city_count = {} # 딕셔너리 생성

for index, row in data.iterrows() :
    city = row['City']
    age = row['Age']

    if city in city_age :
        city_age[city] += age
        city_count[city] += 1
    else :
        city_age[city] = age
        city_count[city] = 1

average = {city: city_age[city] / city_count[city] for city in city_age}
print(average)

for city, age in average.items() :
    print(f"{city} : {age:/2f}")
    

print("가장 많은 사람이 살고 있는 도시 : ")
print(data["City"].value_counts().idxmax())
# most_pop = data['City'].value_counts().idmax()
# most_pop

# 연봉을 추가해서 랜덤으로 할당
# data['Salary'] = np.random.randint(50000, 3000000, size = data.shape[0])
# data.head()

print("\n연봉 순서대로 정렬 : ")
data = data.sort_values("Salary", ascending =  False)
print(data[["Name", "Salary"]])

print("연봉 1등 : ")
print(data.iloc[0]["Name"], data.iloc[0]["Salary"])
# 또는 print(data.loc[data['Salary'].idxmax()])

print("연봉 꼴등:")
print(data.iloc[-1]["Name"], data.iloc[-1]["Salary"])
# 또는 print(data.loc[data['Salary'].idxmin()])

data = data.sort_values('Salary')
print(data.iloc[0])
print(data.iloc[len(data)-1])

print(data.hear(1))
print(data.tail(1))
```
