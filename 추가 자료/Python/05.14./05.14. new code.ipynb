```python
# 데이터 프레임 병합(merge, concat)
df1 = pd.DataFrame({'A':['A'+str(i) for i in range(4)]})
df1
```

```python
# 데이터 프레임 병합(merge, concat)
df2 = pd.DataFrame({'B':['B'+str(i) for i in range(4)]})
df2
```

```python
# 위 두 개의 값을 합쳐준다.
# df1과 df2 병합
# pd.merge(df1, df2)만 써주면 두 데이터 간 공통된 것이 없기 때문에 오류가 발생한다.

df3 = pd.DataFrame({'B':['B'+str(i) for i in range(4)], 'C':['C'+str(i) for i in range(4)]})
df3.loc[4] = ['B4', 'C4']
df3
```

```python
pd.merge(df2, df3)
```

```python
df4 = df2.copy()
df4.loc[3] = 'B0'
df4
```

```python
df5 = df3.copy()
df5.loc[4, 'B'] = 'B0'
df5
```

```python
# df4와 df5의 병합
pd.merge(df4, df5) # 이렇게 작성하면 데이터가 늘어난 모습을 살펴볼 수 있다.
```

```python
pd.merge(df4, df5, how='inner') # default = inner (outer, left, right)
```

```python
pd.merge(df4, df5, how='outer') # outer : 합집합
```

```python
pd.merge(df4, df5, how='left') # left : 왼쪽 데이터 기준
```

```python
pd.merge(df4, df5, how='right') # right : 오른쪽 데이터 기준
```

```python
# key 값을 기준으로 병합
left = pd.DataFrame({'key1':['foo', 'bar', 'bar'],
                     'key2':['one', 'two', 'one'],
                     'lval':[1, 2, 3]})

left
```

```python
right = pd.DataFrame({'key1':['foo', 'foo', 'bar', 'bar'],
                     'key2':['one', 'one', 'one', 'two'],
                     'lval':[4, 5, 6, 7]})

right
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer') # on : 병합키
```

```python
pd.merge(left, right, on='key1') # key1만 가지고 병합
```

```python
# x, y에 대한 이름 직접 지정
pd.merge(left, right, on='key1', suffixes=('_left', '_right'))
```
