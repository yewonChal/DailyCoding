```python
# 정규식을 활용한 bs4 고급 스킬
from bs4 import BeautifulSoup
import re

html = """<html> <head><title>test site</title></head> <body> <div><p id="i"
class="a">test1</p><p class="d">test2</p></div><p class="d">test3</p></p> <a
href="/example/test1">a tag</a> <b>b tag</b></body></html>"""

# 클래스에 d가 포함된 모든 요소 가져오기
# 아이디에 i가 포함된 모든 요소 가져오기
# 태그에 t가 포함된 모든 요소 가져오기 #태그가t로시작하는모든요소가져오기
# href에 슬래시(/)가 포함된 모든 요소 가져오기

soup = BeautifulSoup(html, 'lxml')
print(soup.find_all(class_=re.compile('d')))
print(soup.find_all(id=re.compile('i')))
print(soup.find_all(re.compile('t')))
print(soup.find_all(re.compile('^t')))
print(soup.find_all(href=re.compile('/')))
```

```python
def replace_newline(soup_html) :
    br_to_newlines = soup_html.find_all("br")
    for br_to_newline in br_to_newlines :
        br_to_newline.replace_with("\n")
    return soup_html
```

# 금일숙제 

1) 노래 가사 검색하는 문제 줄바꿈해서 이쁘게 출력하기.

2) 적어주신 URL(연합뉴스 페이지) 속보 기사 제목 출력해보기.


```python
import requests
from bs4 import BeautifulSoup

title = input("제목을 입력하시오. : ")
artist = input("가수를 입력하시오. : ")

url = "https://search.naver.com/search.naver?query=" + artist + " " + title + " 가사"

response = requests.get(url)

if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'lxml')
    lyrics = soup.select_one('div.lyrics_area > div.lyrics_txt')
    if lyrics is None:
        print("찾으시는 곡의 가사 정보를 찾을 수 없습니다.")
    else:
        lines = lyrics.get_text().split('\n')
        for line in lines:
            print(line.strip())
else:
    print(f"Error: {response.status_code}")
```
