```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib import rc
import numpy as np
import seaborn as sns

rc('font', family='AppleGothic')
rc('axes', unicode_minus=False)
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
# 과제1) 
# x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
# x = np.arange(len(x_label))
# y_1 = [66, 80, 60, 50, 80, 10] # john
# y_2 = [55, 90, 40, 60, 70, 20] # peter
# 바그래프 만들기
# -------------

plt.figure(figsize=(10, 5))

x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10] # john
y_2 = [55, 90, 40, 60, 70, 20] # peter

width = 0.3

plt.bar(x - width/2, y_1, width, label='john', color='skyblue')
plt.bar(x + width/2, y_2, width, label='peter', color='orange')

plt.xticks(x, x_label, size = 12)
plt.ylabel('Number of Students')
plt.title('Scores', size = 20)
plt.legend(prop={'size': 12})
plt.show()
```

```python
# 숙제 2
# titanic = sns.load_dataset("titanic")
# 생존여부에 따른 나이와 요금 관계
# -------------

titanic = sns.load_dataset("titanic")

plt.figure(figsize=(10, 5))
sns.scatterplot(data=titanic, x='age', y='fare', hue='survived', palette='RdBu', alpha=0.7)
# 파란색은 생존자를, 빨간색은 그렇지 못한 경우를 나타냅니다.

plt.title('타이타닉호, 나이와 요금에 따른 생존여부', size=15)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
# 강사님 코드
# 과제 1
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10] # john
y_2 = [55, 90, 40, 60, 70, 20] # peter

width = 0.35
plt.rcParams["figure.figsize"] = (8, 4)

fig, axes = plt.subplots()

axes.bar(x - width/2, y_1, width, align='center', alpha=0.5)
axes.bar(x + width/2, y_2, width, align='center', alpha=0.8)

plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Scores')
plt.title('Student')

plt.legend(['john', 'peter'])
plt.show()
```

```python
# 데이터프레임으로 데이터 처리 후 작업
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))
y_1 = [66, 80, 60, 50, 80, 10] # john
y_2 = [55, 90, 40, 60, 70, 20] # peter

data = {
    'Subject' : x_label * 2,
    'Score' : y_1 + y_2,
    'Student' : ['jhon']*len(x_label) + ['peter']*len(x_label)
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 4))
sns.barplot(x='Subject', y='Score', hue='Student', data=df, alpha=0.7)

plt.ylabel('Scores')
plt.title('Students')
plt.show()
```

```python
x = np.arange(2020, 2027)
y1 = np.array([1, 3, 7, 5, 9, 7, 14])
y2 = np.array([1, 3, 5, 7, 9, 11, 13])

fig, ax1 = plt.subplots()

ax1.plot(x, y1, 'g-s', label='Price', alpha = 0.5)
ax1.set_ylim(0, 17)
ax1.set_xlabel('Year')
ax1.set_ylabel('Price')
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.bar(x, y2, color='yellow', label='Demand', alpha=0.3)
ax1.set_ylim(0, 17)
ax2.set_ylabel('Demand')
ax2.tick_params(axis='y', direction='in')

plt.show()
```
