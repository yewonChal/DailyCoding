```python
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Value': np.random.randint(1, 101, size=9),
    'Group': ['X', 'X', 'X', 'Y', 'Y', 'Y', 'Z', 'Z', 'Z']
})

# 1. Category 열을 기준으로 각 카테고리 별 value의 평균을 막대그래프로 나타내라.
mean_ = df.groupby('Category')['Value'].mean().plot(kind='bar', color='pink')
plt.title('Category 열 기준, 각 카테고리 별 value 평균')
plt.xlabel('Category')
plt.ylabel('Average Value')
plt.xticks(rotation = 0)
plt.show()

#2. Group열을 기준으로 각 그룹별 value의 합을 파이 그래프로 나타내라.
sum_ = df.groupby('Group')['Value'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Group열 기준, 각 그룹별 value의 합')
plt.ylabel('') # ylabel에 적힌 'value' 단어 삭제
plt.show()
```

- 스케일링은 데이터의 값을 일정한 범위로 조정하는 작업을 의미한다.
- 주로 0과 1 사이의 범위로 조정하는데 이를 통해 데이터 간의 비교나 모델 학습 등을 효과적으로 진행할 수 있다.

1) 데이터 크기 조정
  
    : 데이터의 범위가 다르면 모델이 학습하기 어려울 수 있다.
  
        - 예를 들어, 한 열의 범위가 0~1, 다른 열의 범위 100~1000이라고 세팅되어 있다면 두 번째 열의 영향을 크게 받는다.
  
        - 따라서 스케일링을 통해 모든 열의 값 범위를 일정하게 조정할 수 있다.
   

3) 알고리즘 안정성 향상

    : 일부 알고리즘은 데이터의 스케일에 민감할 수 있다.
   

5) 속도 향상

    : 일부 최적화 알고리즘에서 최적화하는데 시간이 오래 걸릴 수 있다.
   
         - 미리 스케일링을 수행하고 속도를 향상시킬 수 있다.

스케일링은 주어진 데이터의 범위를 모든 값이 동일한 스케일 내에 있도록 만들어주는 작업이다.

일반적으로 주어진 데이터에서 최솟값을 뺀 후에, 최댓값과 최솟값의 차이로 나누어 줌으로써 스케일링을 수행한다.

```python
data = pd.DataFrame({
    'A' : np.random.randint(0, 100, 10),
    'B' : np.random.randint(0, 100, 10),
    'C' : np.random.randint(0, 100, 10),
    'D' : np.random.randint(0, 100, 10),
    'E' : np.random.randint(0, 100, 10)
})

df_scaled = (data - data.min()) / (data.max() - data.min())

data.plot(kind='bar')
```
