# 박스플롯
: 데이터의 분포를 시각적으로 표현하기 위해 사용되는 그래프(중앙값, 사분위수, 범위, 이상치)

- 중앙값 : 데이터의 중간값, 박스 내부에 가로선으로 나타난다.
- 사분위수 : 제1사분위수(Q1), 하위 25% 보다 크거나 같고 상위 75%보다 작은 값, 박스의 하단 경계선
           제3사분위수(Q3), 하위 25% 보다 크거나 같고 상위 75%보다 작은 값, 박스의 상단 경계선
- 인터사분위 범위(IQR) : Q1과 Q3 사이의 범위를 나타낸다. 중앙값이 포함되며 IQR = Q3 - Q1
- 수염(Whiskers) : 수염은 Q1 이하 Q3 이상의 값을 시각적으로 표현한다. 최대 1.5*IQR 범위 내에 값을 포함한다.

(Q1 - 1.5*IQR, Q3 + 1.5*IQR)

- 이상치 : 수염을 벗어난 값들, 박스플롯에 점으로 나타난다.(분포의 특이점이나 변동성을 표현)

```python
def boxplot_data(spread_scale, center_value, high_offset, low_offset, num_spread=50, num_center=25, num_filer=10):
    spread = np.random.rand(num_spread)*spread_scale
    center = np.ones(num_center)*center_value
    filer_high = np.random.rand(num_filer)*spread_scale+high_offset
    filer_low = np.random.rand(num_filer)*low_offset
    data = np.concatenate((spread, center, filer_high, filer_low))
    return data

data1 = boxplot_data(100, 50, 100, -100)
data2 = boxplot_data(100, 40, 100, -100)

data = [data1, data2, data2[::2]]

plt.boxplot(data)
plt.show()
```

```python
plt.boxplot(data, vert=False)
plt.show()
```

```python
outlier_marker = dict(markerfacecolor='r', marker='d')

plt.boxplot(data, flierprops=outlier_marker)
plt.show()
```

```python
ax = plt.axes(projection='3d')

x = np.cumsum(np.random.normal(0, 1, 100))
y = np.cumsum(np.random.normal(0, 1, 100))
z = np.cumsum(np.random.normal(0, 1, 100))

ax.plot3D(x, y, z, alpha=0.5, marker='x')

plt.title('plot 3D')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d') # 1행, 1열, 첫번째 인덱스로 선택하겠다.
# c = z : z값에 따라 색깔을 다르게 설정하겠다.
# s : size점의 크기를 20으로 지정하겠다.
# cmap : 색상 맵을 Green으로 지정하겠다.
ax.scatter(x, y, z, c=z, s=20, alpha=0.5, cmap='Greens')
plt.title("3d scatter")
plt.show()
```

```python
# 등고선
x = np.linspace(-8, 8, 40)
y = np.linspace(-8, 8, 40)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection='3d')

ax.contour3D(x, y, z, 20, cmap='Reds')
plt.title('contour 3D')
plt.show()
```
