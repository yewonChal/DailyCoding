```python
import seaborn as sns # 통계적 데이터분석에 특화되어 있는 라이브러리
```

```python
tips = sns.load_dataset('tips') # sns에서 제공하는 미국 ?? 레스토랑의 수집된 팁과 관련된 데이터셋
tips.head()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('violin plot')
plt.show()
```

```python
sns.countplot(x='day', data=tips)
plt.title('countplot')
plt.show()
```

```python
sns.relplot(x='tip', y='total_bill', data=tips)
plt.title('replot')
plt.show()
```

```python
sns.lmplot(x='tip', y='total_bill', data=tips)
plt.title('lmplot')
plt.show()
```

```python
numeric_tips = tips.select_dtypes(include=['number'])
numeric_tips.corr() # 상관관계 출력
```

```python
sns.heatmap(numeric_tips.corr(), annot=True, linewidths=1, cmap='coolwarm')
plt.title('heatmap')
plt.show()
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips) # , palette='colorblind'
plt.show()
```

```python
# HUSL 색상 공간을 사용해서 밝은 팔레트 생성
sns.palplot(sns.light_palette((210, 90, 60), input='husl'))

# XKCD 색상 공간을 사용해서 어두운 팔레트 생성
sns.palplot(sns.dark_palette('muted purple', input='xkcd'))

# ColorBrewer 팔레트에서 BrBG(브라운 - 그린) 팔레트 생성
sns.palplot(sns.color_palette('BrBG', 10))

# ColorBrewer 팔레트에서 BrBG_r(브라운 - 그린 반전) 팔레트 생성
sns.palplot(sns.color_palette('BrBG_r', 10))

sns.palplot(sns.color_palette('coolwarm', 10))

# 중심이 어두운 색인 발산형 팔레트 생성
sns.palplot(sns.diverging_palette(255, 133, l=60, n=10, center='dark'))

# 색상 url : https://seaborn.pydata.org/generated/seaborn.color_palette.html
```

```python
sns.barplot(x='tip', y='total_bill', data=tips, palette='pastel')
```

```python
sns.catplot(x='sex', y='total_bill', data=tips, kind='bar')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', data=tips, kind='bar')
plt.show()
```

```python
sns.catplot(x='sex', y='total_bill', hue='smoker', col='time', data=tips, kind='bar')
plt.show()
```

```python
# c, color
# matplot에선 주로 color를 사용한다.
# seaborn에선 주로 c를 사용한다.
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.figure(figsize=(8,4))
plt.bar(x, y, align='center', color='red', alpha=0.7)

plt.xticks(x)
plt.ylabel('Scores')
plt.title('Subjects')
plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

sns.barplot(x=x, y=y, alpha=0.7, palette='YlGnBu')

plt.ylabel('Scores')
plt.title('Subjects')
plt.show()
```

```python
x = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.figure(figsize=(8,4))
plt.barh(x, y, align='center', color='purple', alpha=0.3)

plt.yticks(x)
plt.ylabel('Scores')
plt.title('Subjects')
plt.show()
```
