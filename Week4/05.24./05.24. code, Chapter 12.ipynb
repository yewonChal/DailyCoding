```python
# 예제 12-5 : 그래프 선 스타일 설정
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family='AppleGothic')
font1 = {'size':20, 'color':'black'}

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
ydata3 = [85, 97, 78]
ydata4 = [92, 88, 82]

plt.plot(xdata, ydata1, label='국어', color='red', linestyle='-', marker='o')
plt.plot(xdata, ydata2, label='영어', color='#00ffff', linestyle='--', marker='x')
plt.plot(xdata, ydata3, label='수학', color='magenta', linestyle='-.', marker='s')
plt.plot(xdata, ydata4, label='사회', color='#444444', linestyle=':', marker='d')

plt.title('[ 세명 학생의 네 과목 성적 ]', fontdict = font1)
plt.legend(loc='best')

plt.show()
```

```python
# 예제 12-6 : X/Y축 범위와 눈금 설정
import matplotlib.pyplot as plt

xdata = list(range(30)) # 0~29
ydata = [] # 빈 리스트 생성

for x in xdata: # 반복문
    y = 2*x # y에 2*x
    ydata.append(y) # ydata에 y값 추가

plt.plot(xdata, ydata, label='y=2x') # x축에 xdata, y축에 ydata, label은 y=2x

plt.xlim(0, 35) # x축 길이를 0~35까지 출력 
plt.ylim(0, 70) # y축 길이를 0~70까지 출력

plt.xticks(list(range(0, 35, 2))) # 2단위씩 증가
plt.yticks(list(range(0, 70, 5))) # 5단위씩 증가

plt.grid(True) # 눈금 입력
plt.show() # 그래프 출력
```

```python
# 예제 12-10 : 다수의 그래프 그리기
x = ['월', '화', '수', '목', '금', '토', '일']
y1 = [6.5, 5.7, 5.5, 6.7, 6.3, 7.5, 8.3]
y2 = [6.3, 7.7, 7.5, 7.7, 6.2, 7.3, 8.5]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(9, 3), sharex=True, sharey=True)
# 칸을 나누고 어떻게 집어넣을 것인지 설명해주는 코드(subplots)
# sharex=True : x축을 공유
# sharey=True : y축을 공유
# fig -> 타이틀
# axs -> 그래프

ax = axs[0]
ax.scatter(x,y1)
ax.set_title('2018')

ax = axs[1]
ax.scatter(x,y2)
ax.set_title('2019')

fig.suptitle('[ 연간 요일별 평균 수면시간 ]', size = 15)
plt.show()
```

```python
# 예제 12-11 : 그래프 간격 조정하기
x = list(range(1,11)) # 1~10
y = list(range(10, 101, 10)) # 10~100, 10 단위로

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(9, 5), sharex=True, sharey=True)
# fig = title
# axs = 그래프
# subplots 칸을 나누고 집어넣는 코드

ax = axs[0][0]
ax.plot(x,y)
ax.set_title('선 그래프 1')

ax = axs[0][1]
ax.plot(x,y, color='red', linestyle='--', marker='o')
ax.set_title('선 그래프 2')

ax = axs[1][0]
ax.bar(x,y)
ax.set_title('막대 그래프')

ax = axs[1][1]
ax.scatter(x,y)
ax.set_title('산포 그래프')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.2, hspace=0.3)
# subplots_adjust : 서브플롯 간 사이즈 조절
# left, right 등은 각 지칭하는 위치(공간)의 여백을 의미한다.

plt.show()
```

```python
# 원 표시에 도시 이름을 적어주자.
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] 
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den)*0.4
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y']

plt.scatter(lon, lat, s=size, c=colors, alpha=0.5) # alpha : 투명도
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('pop_den')

for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)

plt.ylim(34, 39)

plt.show()
```
