```python
# 위 코드는 오류났던 코드, 밑 코드가 강사님께서 작성하신 코드
# 네이버 지도 접속 - 검색창에 '강남 카페' 검색 - 장소 칸 스크롤 내리기 - 다음 페이지(p.2)로 이동
# - 다시 장소 칸 스크롤 내리기 - 조회한 장소의 가게명 추출하기
driver = webdriver.Chrome() # Chrome에서 페이지를 열 수 있도록 설정
driver.get('https://www.naver.com') # 접속 url

original_window_handle = driver.current_window_handle
# 현재 창의 핸들을 저장

driver.find_element(By.CSS_SELECTOR, 'span.type_map').click() # 접속한 url의 페이지 소스 'span.type_map'을 클릭(CSS_SELECTOR)
time.sleep(2) # 딜레이(접속에 텀을 두어 안전하게 접근하기 위해)

new_window_handle = None # 새로운 창 변수에 None값(없는 값)을 넣어준다.
for handle in driver.window_handles: # new_window_handle = driver.window_handles[1]
# 반복문, handle이 driver.window_handles 즉, new_window_handle 내에 존재할 때까지 반복
# 정리하자면, 새로운 창을 찾기 전까지 반복하는 코드
    if handle != original_window_handle:
    # 만약 handle의 값이 original_window_handle(기존에 켰던 창)과 같지 않다면,
        new_window_handle = handle
        # new_window_handle에 그 값(handle값)을 넣어준다.
        break
        # 이후 break을 통해 반복문을 종료시켜준다.

driver.switch_to.window(new_window_handle)
# driver에 switch_to.window(new_window_handle)을 통해
# 위 반복문과 조건문에서 받아온 새로운 창을 띄워준다.
# = 새로운 창으로 전환

try: # 예외처리
    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'input_search'))
        # 검색 입력창이 로드될 때까지 10초 대기(안전성을 위해(파파고 예제 참고))
    )
except:
    print('해당 검색 셀렉터가 없습니다.') # 문장을 출력한다.(예외 발생 시)
finally:
    pass # 다음 코드로 넘어간다.

search = driver.find_element(By.CLASS_NAME, 'input_search')
# search 변수에 새로 접속한 창의 소스(속성) 'input_search'를 입력해준다.
search.send_keys('강남 카페\n')
# 위 입력된 칸에 '강남 카페'를 입력하고 '\n'을 통해 검색을 진행한다.('\n' = enter key)
#search.send_keys(Keys.ENTER)
time.sleep(2) # 딜레이(안전하게 접근하기 위해)

# places = driver.find_element(By.CLASS_NAME, 'panel_content')
# print(places.text)

driver.switch_to.frame("searchIframe")

try: # 예외처리
    element = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'place_bluelink'))
    )
finally:
    pass

driver.find_element(By.ID, '_pcmap_list_scroll_container').click()
# driver 페이지 소스 중 '_pcmap_list_scroll_container'값을 클릭해준다.

list_len = driver.find_elements(By.CSS_SELECTOR, 'li.UEzoS')
# list_len에 driver의 페이지 소스 중 'li.UEzoS'을 선택하는 값을 입력해준다.
before_len = len(list_len)
# before_len에 list_len 길이값을 입력해준다.
  
while True:  
    driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
    time.sleep(0.5) # 딜레이
    
    list_len = driver.find_elements(By.CSS_SELECTOR, 'li.UEzoS')
    # list_len에 driver의 페이지 소스 중 'li.UEzoS'을 선택하는 값을 입력해준다.
    after_len = len(list_len) # after_len에 list_len의 길이값을 입력해준다.
    
    # if before_len < after_len:
    #     print("스크롤 되었습니다.")
    #     before_len = after_len
    # else:
    #     print("페이지 끝에 도달했습니다.")
    #     break
    
    if before_len == after_len:
    # 만약 before_len값과 after_len값이 같다면, break를 걸어준다.
        break
        
    before_len = after_len

# Before : 
# After : 
# 각 값을 출력해보면 이해하기 쉽다.
    
places = driver.find_elements(By.CLASS_NAME, 'place_bluelink')
# places에 driver

print(1, '----------------------------------------')
for place in places:
    print(place.text)

driver.find_element(By.XPATH, '//a[text()=2]').click()
time.sleep(1)

while True:  
    driver.find_element(By.CSS_SELECTOR, 'body').send_keys(Keys.END)
    time.sleep(0.5)
    
    list_len = driver.find_elements(By.CSS_SELECTOR, 'li.UEzoS')
    after_len = len(list_len)
    
    # if before_len < after_len:
    #     print("스크롤 되었습니다.")
    #     before_len = after_len
    # else:
    #     print("페이지 끝에 도달했습니다.")
    #     break
    
    if before_len == after_len:
        break
        
    before_len = after_len
    
places = driver.find_elements(By.CLASS_NAME, 'place_bluelink')

print(2, '----------------------------------------')
for place in places:
    print(place.text)

time.sleep(5)
driver.quit()
```

```python
# https://pixabay.com/ 여기에서 이미지 다운로드
import os

driver = webdriver.Chrome()

search_keywords = ['dog', 'cat', 'bird']

for keyword in search_keywords:
    driver.get('https://pixabay.com/')

    time.sleep(2)

    search_box = driver.find_element(By.NAME, 'search')
    search_box.send_keys(keyword)
    search_box.submit()
    time.sleep(2)

    image_links = []
    while len(image_links) < 3 :
        images = driver.find_elements(By. TAG_NAME, 'img')

        for image in images:
            src = image.get_attribute('src')
            if src and 'https://cdn.pixabay.com/' in src: # 무료 이미지만 검색하기 위해
                image_links.append(src)

        image_links = list(set(image_links))
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(2)

    image_links = image_links[:5]

    # 이미지 다운로드 작업
    save_dir = f'images/{keyword}'
    os.makedirs(save_dir, exist_ok=True)

    for i, link in enumerate(image_links):
        response = requests.get(link)
        with open(f'{save_dir}/image{i+1}.jpg', 'wb') as f : # 이미지는 바이너리 작업 필수
            f.write(response.content)

driver.quit()
```
